```python
import os
import pandas as pd
```

```python
print(os.listdir("dataset"))
```

```python
F2000_dir = "dataset/MICC-F2000"
F2000 = os.listdir(F2000_dir)
print(len(F2000))


file_paths = []
labels = []

for filename in F2000:
    if "tamp" in filename:
        file_paths.append(os.path.join(F2000_dir, filename))
        labels.append("tampered")
    elif "scale" in filename:
        file_paths.append(os.path.join(F2000_dir, filename))
        labels.append("original")

data = {'Path': file_paths, 'Label': labels}
F2000 = pd.DataFrame(data)

F2000.head()
```

```python
F2000["Label"].value_counts()
```

```python
F600_dir = "dataset/MICC_F600"
F600 = os.listdir(F600_dir)
print(len(F600))

### THERE'S A PROBLEM HERE
file_paths = []
labels = []

for filename in F600:
    if "_scale" not in filename and "gt" not in filename:
        file_paths.append(os.path.join(F600_dir, filename))
        labels.append("tampered")
    elif "_scale" in filename and "gt" not in filename:
        file_paths.append(os.path.join(F2000_dir, filename))
        labels.append("original")

data = {'Path': file_paths, 'Label': labels}
F600 = pd.DataFrame(data)

print(len(F600))

F600.head()
```

```python
F600["Label"].value_counts()
```

```python
F8multi_dir = "dataset/MICC-F8multi"
F8multi = os.listdir(F8multi_dir)
print(len(F8multi))

F8multi_ = pd.DataFrame()

file_paths = []
labels = []

for filename in F8multi:
    file_paths.append(os.path.join(F8multi_dir, filename))
    labels.append("tampered")

F8multi_["Path"] = file_paths
F8multi_["Label"] = labels

F8multi_.head()
```

```python
F8multi_["Label"].value_counts()
```

```python
F220_dir = "dataset/MICC-F220"
F220 = os.listdir(F220_dir)
print(len(F220))


file_paths = []
labels = []

for filename in F220:
    if "tamp" in filename:
        file_paths.append(os.path.join(F220_dir, filename))
        labels.append("tampered")
    elif "scale" in filename:
        file_paths.append(os.path.join(F220_dir, filename))
        labels.append("original")

data = {'Path': file_paths, 'Label': labels}
F220 = pd.DataFrame(data)

F220.head()
```

```python
F220["Label"].value_counts()
```
